```python
import json
import numpy as np
import pandas as pd
```

```python
#Load all of the ingredient data
with open('recipe-ingredients-dataset/train.json') as jfile:
    data = json.load(jfile)
```

```python
#understand the format of the data
print(data[0])
```

```python
#create a dictionary with a count of the mentions of ingredients
bag_of_words = {}
for example in data:
    for ingredient in example['ingredients']:
        bag_of_words[ingredient] = bag_of_words.get(ingredient,0) + 1
        
print(bag_of_words)
```

```python
'''
remove all items from the dictionary if they have been mentioned less than 30 times for ease. These items will be 
replaced with N/A for training to ease the training process
'''
del_list = []
for item in bag_of_words:
    if bag_of_words[item] < 30:
        del_list.append(item)
for item in del_list:
    del bag_of_words[item]
```

```python
print(len(bag_of_words))
```

```python
#Function using above discovery to create a dataframe with the bag of words, and a list of the keys of the dataframe
def create_food_bag_of_words(file_path = 'recipe-ingredients-dataset/train.json', cuttoff_threshold = 30):
    with open(file_path) as jfile:
        data = json.load(jfile)
    print('There are {} recipe examples in this dataset.'.format(len(data)))
    bag_of_words = {}
    for example in data:
        for ingredient in example['ingredients']:
            bag_of_words[ingredient] = bag_of_words.get(ingredient,0) + 1
    print('There are {} ingredients in this dataset.'.format(len(bag_of_words)))
    del_list = []
    for item in bag_of_words:
        if bag_of_words[item] < cuttoff_threshold:
            del_list.append(item)
    for item in del_list:
        del bag_of_words[item]
    print('There are {} ingredients in this dataset mentioned more than 30 times.'.format(len(bag_of_words)))
    return(pd.Series(bag_of_words), list(bag_of_words.keys()))
    
```

```python
bag_of_words, keys = create_food_bag_of_words()
```
